```python
import numpy as np
import plotly.graph_objects as go


# Define Gaussian log-likelihood function
def gaussian_log_likelihood(mu, sigma, x):
    return -0.5 * np.sum(((x - mu) / sigma) ** 2 + np.log(2 * np.pi * sigma**2))


# Generate some example data points
np.random.seed(42)
x = np.random.normal(5, 2, 100)

# Define a range of standard deviations (sigma) and a fixed mean (mu)
sigma_range = np.linspace(0.5, 5, 100)  # Sigma from 0.5 to 5
mu = 5  # Fixing mu at 5

# Calculate log-likelihood for each sigma
log_likelihoods = [gaussian_log_likelihood(mu, sigma, x) for sigma in sigma_range]

# Create an interactive plot using Plotly
fig = go.Figure()

# Add the Gaussian log-likelihood curve (log-likelihood vs sigma)
fig.add_trace(
    go.Scatter(
        x=sigma_range,
        y=log_likelihoods,
        mode="lines",
        name="Log Likelihood",
        hoverinfo="x+y",  # Show x and y values on hover
        line=dict(color="royalblue", width=3),
    )
)

# Customize layout
fig.update_layout(
    title="Gaussian Log-Likelihood vs Sigma",
    xaxis_title="Standard Deviation (σ)",
    yaxis_title="Log-Likelihood",
    hovermode="closest",
)  # Show info of closest point on hover

# Show the plot
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot

# Initialize notebook mode for offline plotting
init_notebook_mode(connected=True)


# Define Gaussian log-likelihood function
def gaussian_log_likelihood(mu, sigma, x):
    return -0.5 * np.sum(((x - mu) / sigma) ** 2 + np.log(2 * np.pi * sigma**2))


# Generate some example data points
np.random.seed(42)
x = np.random.normal(5, 2, 100)

# Define a range of means (mu) and a fixed sigma (standard deviation)
mu_range = np.linspace(0, 10, 100)
sigma = 2  # Fixing sigma

# Calculate log-likelihood for each mu
log_likelihoods = [gaussian_log_likelihood(mu, sigma, x) for mu in mu_range]

# Create an interactive plot using Plotly
fig = go.Figure()

# Add the Gaussian log-likelihood curve
fig.add_trace(
    go.Scatter(
        x=mu_range,
        y=log_likelihoods,
        mode="lines",
        name="Log Likelihood",
        hoverinfo="x+y",  # Show x and y values on hover
        line=dict(color="royalblue", width=3),
    )
)

# Customize layout
fig.update_layout(
    title="Gaussian Log-Likelihood",
    xaxis_title="Mean (μ)",
    yaxis_title="Log-Likelihood",
    hovermode="closest",
)

# Show the plot in Jupyter
iplot(fig)
```
